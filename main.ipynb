# Tìm Hiểu Về Cronbach's Alpha Từ Dữ Liệu Flash Sale Shopee

---

## 1. Bắt đầu từ một tình huống rất quen với marketer

Giả sử bạn muốn trả lời một câu hỏi rất thực tế:

> **Flash sale có thật sự làm khách hàng thấy gấp, thấy thích thú, hoặc dễ mua ngẫu hứng hơn không?**

Nếu chỉ hỏi một câu, kết quả thu được thường không đủ để phản ánh đầy đủ khái niệm cần đo và dễ bị ảnh hưởng bởi cách diễn đạt của riêng câu hỏi đó. Vì vậy, trong nghiên cứu marketing, người ta thường dùng **một nhóm câu hỏi** để đo cùng một nội dung.

Ví dụ:

- để đo **cảm giác cấp bách**, có thể hỏi nhiều câu về việc sợ hết hàng, sợ bỏ lỡ giá tốt, phải quyết định nhanh,
- để đo **sự thích thú**, có thể hỏi nhiều câu về cảm giác vui vẻ, hạnh phúc, phấn khích.

Khi đã có một nhóm câu hỏi như vậy, ta cần kiểm tra xem chúng có **đủ nhất quán với nhau** để gộp lại thành một thang đo hay không. Đó chính là lúc dùng **Cronbach's Alpha**.

---

## 2. Cronbach's Alpha là gì?

Cronbach's Alpha là một chỉ số giúp trả lời câu hỏi sau:

> **Các câu hỏi trong cùng một nhóm có mức độ nhất quán với nhau hay không?**

Hiểu đơn giản:

- **Alpha cao**: các câu hỏi khá ăn khớp với nhau và cho thấy mức độ nhất quán nội tại tốt trên mẫu hiện tại.
- **Alpha thấp**: các câu hỏi chưa hợp nhau lắm, có thể có câu bị lệch ý hoặc đang đo sang phần khác.

Bạn có thể hiểu Cronbach's Alpha như một cách kiểm tra **độ nhất quán nội tại** của thang đo. Nói dễ hiểu hơn, nó cho biết các câu hỏi trong cùng một nhóm có mức độ nhất quán với nhau đủ tốt để được xem xét như một thang đo chung hay không.

Điểm quan trọng cần nhớ:

- Cronbach's Alpha chủ yếu cho biết **mức độ nhất quán giữa các câu hỏi trong cùng một thang đo**.
- Nó **không đủ để kết luận** thang đo là **hợp lệ**.
- Nó cũng **không tự khẳng định** rằng các item **chỉ đo một khái niệm duy nhất**; điều đó cần được kiểm tra thêm bằng **Exploratory Factor Analysis (EFA)**.
- Ngay cả khi alpha cao, các item từ nhiều construct khác nhau vẫn có thể có tương quan trong một bối cảnh cụ thể; nếu gộp chúng lại, composite score trở nên mơ hồ về mặt khái niệm.
- Nếu muốn nhìn kỹ từng câu hỏi, cần xem thêm các chỉ số như `Corrected Item-Total Correlation` và `Cronbach's Alpha if Item Deleted`.
- Kết quả này luôn phải được hiểu **trên mẫu dữ liệu hiện tại**, không nên xem như một nhãn đúng cho mọi mẫu dữ liệu khác.

Cần lưu ý rằng Cronbach's Alpha phản ánh độ nhất quán nội tại giữa các item, chứ không tự chứng minh rằng các item chỉ phản ánh một nhân tố duy nhất hay thang đo hoàn toàn hợp lệ về mặt khái niệm. Vì vậy, sau bước kiểm tra Cronbach's Alpha, nghiên cứu vẫn cần sử dụng **Exploratory Factor Analysis (EFA)** để xem các item có thực sự hội tụ vào đúng các khái niệm dự kiến hay không.

## 3. Ví dụ thật từ bộ dữ liệu này

### Ví dụ 1: Thang đo `Urgency`

Trong bộ dữ liệu này, thang đo `Urgency` được đo bằng 5 câu hỏi thật như sau:

- `Urgency_1`: "Tôi cho rằng mình cần phải mua ngay lập tức để sở hữu được sản phẩm này."
- `Urgency_2`: "Tôi cho rằng nếu không mua ngay, sản phẩm sẽ sớm bị bán hết."
- `Urgency_3`: "Tôi cho rằng mình phải quyết định mua nhanh trước khi Flash Sale kết thúc."
- `Urgency_4`: "Tôi cho rằng nếu không mua ngay, tôi có thể bỏ lỡ cơ hội sở hữu sản phẩm với giá ưu đãi."
- `Urgency_5`: "Tôi cho rằng mình sẽ hối tiếc nếu không mua ngay lập tức."

Nếu một người thực sự cảm thấy cấp bách, họ thường sẽ chọn điểm khá cao ở nhiều câu trong nhóm này. Khi đó, các câu hỏi được xem là khá ăn khớp với nhau.

Trong dữ liệu hiện tại, thang đo `Urgency` có Cronbach's Alpha = `0.876`, tức là **độ nhất quán nội tại tốt**.

### Ví dụ 2: Thang đo `Enjoyment`

Thang đo `Enjoyment` trong bộ dữ liệu này gồm 5 câu hỏi:

- `Enjoyment_1`: Flash Sale khiến tôi cảm thấy vui vẻ.
- `Enjoyment_2`: Flash Sale khiến tôi cảm thấy hạnh phúc.
- `Enjoyment_3`: Flash Sale khiến tôi cảm thấy phấn khích.
- `Enjoyment_4`: Flash Sale khiến tôi cảm thấy thích thú.
- `Enjoyment_5`: Flash Sale khiến tôi giảm bớt cảm giác mệt mỏi.

Nếu người trả lời chọn cao ở hầu hết các câu này, điều đó cho thấy nhóm câu hỏi này có mức độ nhất quán nội tại khá tốt trên mẫu hiện tại, dù vẫn có thể cần kiểm tra thêm từng item ở bước `Corrected Item-Total Correlation`, `Cronbach's Alpha if Item Deleted`, hoặc Exploratory Factor Analysis (EFA).

Trong dữ liệu hiện tại, thang đo `Enjoyment` có Cronbach's Alpha = `0.902`, tức là **độ nhất quán nội tại rất tốt**.

### Ví dụ 3: Một trường hợp alpha thấp (ví dụ minh họa)

Ngoài ví dụ từ dữ liệu thật, ta cũng nên xem một ví dụ giả định về trường hợp alpha thấp để dễ phân biệt.

Đây là **ví dụ giả định**, không phải kết quả thật từ bộ dữ liệu Shopee Flash Sale.

Giả sử ta muốn đo "sự thích thú khi mua flash sale", nhưng lại gom các câu hỏi sau vào cùng một nhóm:

- "Tôi thấy mua flash sale khá vui."
- "Tôi cảm thấy hào hứng khi tham gia flash sale."
- "Flash sale làm trải nghiệm mua sắm thú vị hơn."
- "Tôi thường đọc đánh giá sản phẩm rất kỹ trước khi mua."
- "Tôi quan tâm đến tốc độ giao hàng hơn là giá khuyến mãi."

Trong nhóm này, 3 câu đầu đang nói về **cảm xúc thích thú**, nhưng 2 câu sau lại nghiêng về **thói quen cân nhắc khi mua hàng**.

Vì các câu hỏi trong nhóm này không đủ nhất quán để hoạt động như một thang đo chung, Cronbach's Alpha có thể xuống thấp, ví dụ chỉ khoảng `0.55`.

Khi gặp trường hợp như vậy, nhà nghiên cứu nên xem lại:

- các câu hỏi có đang đo cùng một nội dung hay không,
- có câu nào bị lệch ý so với phần còn lại hay không,
- có nên ghi nhận hạn chế hoặc cân nhắc loại bớt câu hỏi khỏi thang đo hiện tại hay không.

### Ví dụ 4: Câu hỏi đảo chiều nếu không xử lý đúng sẽ làm alpha thấp đi

Đây cũng là **ví dụ giả định**, nhưng rất hay gặp trong nghiên cứu thực tế.

Giả sử bạn có 3 câu cùng đo cảm giác cấp bách:

- "Tôi thấy cần mua ngay khi có flash sale."
- "Tôi sợ hết hàng nếu không mua ngay."
- "Tôi không thấy gấp chút nào khi có flash sale."

Một người thật sự thấy rất cấp bách có thể trả lời như sau:

- câu 1: `5`
- câu 2: `5`
- câu 3: `1`

Về mặt ý nghĩa, câu trả lời này rất hợp lý. Người đó đồng ý mạnh với 2 câu đầu và không đồng ý với câu thứ 3 vì câu thứ 3 đang viết theo hướng ngược lại.

Nhưng nếu giữ nguyên dữ liệu là `5, 5, 1`, phần mềm sẽ dễ hiểu rằng câu thứ 3 đang đi ngược với 2 câu còn lại. Khi đó Cronbach's Alpha có thể bị kéo xuống thấp một cách oan uổng.

Lúc này cần **đảo mã** câu hỏi ngược chiều trước khi tính alpha:

- `1` đổi thành `5`
- `2` đổi thành `4`
- `3` giữ nguyên `3`
- `4` đổi thành `2`
- `5` đổi thành `1`

Sau khi đảo mã, bộ điểm `5, 5, 1` sẽ trở thành `5, 5, 5`. Khi đó phần mềm mới hiểu đúng rằng cả 3 câu đều đang phản ánh cùng một mức độ cấp bách.

Vì vậy, khi thấy alpha thấp, đừng vội kết luận thang đo kém. Trước tiên nên kiểm tra xem có câu hỏi đảo chiều mà chưa được xử lý hay không.

### Ví dụ 5: Nếu mẫu quá nhỏ thì alpha gần như không nói được nhiều

Đây cũng là một tình huống rất quan trọng nhưng dễ bị bỏ qua.

Giả sử bạn có một thang đo gồm 5 câu hỏi, nhưng chỉ có **1 người** trả lời khảo sát.

Người đó trả lời như sau:

- câu 1: `5`
- câu 2: `5`
- câu 3: `4`
- câu 4: `5`
- câu 5: `4`

Nhìn bằng mắt thường, bạn có thể thấy người này trả lời khá nhất quán. Nhưng vấn đề là **Cronbach's Alpha không được tạo ra để đọc từ 1 người trả lời**.

Lý do là alpha cần nhìn sự biến thiên giữa **nhiều người** để xem các câu hỏi có cùng thay đổi với nhau hay không. Nếu chỉ có 1 dòng dữ liệu, gần như không có đủ thông tin để tính phương sai và hiệp phương sai một cách có ý nghĩa.

Nói dễ hiểu hơn:

- `1 người trả lời` thì không thể kết luận thang đo tốt hay không tốt bằng Cronbach's Alpha,
- `rất ít người trả lời` thì dù phần mềm có tính ra một con số, con số đó cũng rất dễ bất ổn,
- mẫu càng đủ lớn và đa dạng, việc diễn giải alpha mới càng đáng tin hơn.

Vì vậy, khi đọc alpha, luôn nên tự hỏi thêm một câu:

> **Con số này được tính trên một mẫu đủ lớn để tin được hay chưa?**

### Ví dụ 6: Alpha cao nhưng không nên gộp — khi items đến từ nhiều construct khác nhau

Đây là **ví dụ giả định**, nhưng phản ánh một rủi ro thực tế thường gặp trong nghiên cứu hành vi tiêu dùng.

Giả sử nhà nghiên cứu muốn tóm tắt "tác động tâm lý tổng thể của flash sale" bằng một điểm số chung, và gộp các item sau vào cùng một thang đo:

- "Tôi cảm thấy cần phải mua ngay." (đo Urgency — cảm giác cấp bách)
- "Tôi cảm thấy phấn khích khi thấy flash sale." (đo Enjoyment — cảm xúc tận hưởng)
- "Tôi thấy mức giảm giá rất hấp dẫn." (đo Eco_Benefit — lợi ích kinh tế)

Trong bối cảnh flash sale, ba item này có thể có tương quan khá tốt với nhau: người cảm thấy gấp thường cũng cảm thấy phấn khích, và cũng thấy giá hấp dẫn. Kết quả là Cronbach's Alpha hoàn toàn có thể đạt `0.78–0.82` — trông có vẻ chấp nhận được.

Tuy nhiên, nếu gộp chúng lại thành một điểm số chung, câu hỏi đặt ra là: **điểm đó đang đo cái gì?**

- Không phải Urgency thuần túy (vì có item về cảm xúc và giá).
- Không phải Enjoyment thuần túy (vì có item về cấp bách và giá).
- Không phải Eco_Benefit thuần túy (vì có item về cảm xúc và áp lực thời gian).

Thang đo trở nên mơ hồ về mặt khái niệm, dù alpha vẫn cao. Đây chính là ranh giới giữa hai câu hỏi khác nhau:

| Công cụ | Câu hỏi đặt ra |
|---|---|
| **Cronbach's Alpha** | Các item này có tương quan đủ tốt với nhau không? |
| **Exploratory Factor Analysis (EFA)** | Chúng có thực sự phản ánh cùng một nhân tố không? |

Alpha cao là điều kiện **cần**, nhưng chưa **đủ**. Nếu EFA cho thấy các item tải lên nhiều nhân tố khác nhau, thì gộp chúng lại thành một thang đo là không phù hợp về mặt học thuật, dù alpha trông ổn.

Nói ngắn gọn: **alpha cao cho phép tính composite score về mặt số học, nhưng không đảm bảo rằng composite score đó có ý nghĩa về mặt khái niệm**.

## 4. Cách đọc kết quả

Các ngưỡng dưới đây chỉ mang tính hướng dẫn tham khảo; việc diễn giải vẫn cần đặt trong bối cảnh nghiên cứu, nội dung item và kết quả phân tích nhân tố.

| Giá trị alpha | Có thể hiểu nhanh | Nên làm gì? |
|---|---|---|
| > 0.95 | Cần kiểm tra lại | Có thể các item đang quá giống nhau và lặp ý; xem lại nội dung để tránh item trùng lặp |
| 0.90-0.95 | Rất tốt | Thường có thể dùng tiếp. Nếu muốn kiểm tra kỹ hơn, hãy xem có câu nào yếu hơn phần còn lại không |
| 0.80-0.89 | Tốt | Có thể dùng tiếp. Nếu muốn chắc hơn, hãy xem có câu nào lệch với các câu còn lại không |
| 0.70-0.79 | Chấp nhận được | Thường vẫn dùng được, nhưng nên đọc kỹ từng câu hỏi trước khi đi tiếp |
| 0.60-0.69 | Hơi yếu | Chưa nên yên tâm, cần xem lại nội dung câu hỏi, cách mã hóa và câu đảo chiều |
| < 0.60 | Thấp | Không nên dùng ngay, cần kiểm tra lại thang đo trước |

Nếu muốn kiểm tra kỹ hơn sau khi nhìn alpha, có 4 bước nên làm theo thứ tự:

1. **Đọc lại nội dung item**: xem có câu nào bị viết lệch ý hoặc quá giống câu khác không.
2. **Xem `Corrected Item-Total Correlation`**: biết có câu nào không đi cùng phần còn lại của thang đo không.
3. **Xem `Cronbach's Alpha if Item Deleted`**: biết nếu bỏ một câu đi thì cả thang đo có tốt hơn rõ rệt không.
4. **Cân nhắc lý thuyết trước khi quyết định loại item**: nếu thống kê gợi ý loại một item, cần xem thêm tầm quan trọng khái niệm của item đó — không nên loại chỉ vì số liệu nếu item đó là cốt lõi của construct.

Ví dụ:

- Nếu `Urgency = 0.876`, có thể nói thang đo này có độ nhất quán nội tại tốt và có thể tiếp tục sử dụng.
- Nếu `Enjoyment = 0.902`, có thể nói thang đo này có độ nhất quán nội tại rất tốt. Nếu muốn kiểm tra kỹ hơn, hãy xem có câu nào yếu hơn phần còn lại hay không.
- Nếu một thang đo chỉ đạt `0.58`, chưa nên dùng ngay. Trước tiên hãy kiểm tra lại nội dung câu hỏi và cách mã hóa.

---

## 5. Khi nào nên dùng Cronbach's Alpha?

Cronbach's Alpha phù hợp khi:

- một nội dung được đo bằng **nhiều câu hỏi Likert**,
- các câu hỏi được **dự kiến tổng hợp thành một điểm số chung** cho cùng một thang đo,
- các câu hỏi được mã hóa theo **cùng một chiều** (hoặc đã đảo mã câu ngược chiều).

Cronbach's Alpha không phù hợp khi:

- chỉ có một câu hỏi,
- đó là thông tin mô tả mẫu như giới tính, tuổi, nghề nghiệp,
- các câu hỏi trong cùng nhóm thực ra đang đo nhiều nội dung khác nhau,
- số quan sát quá ít để phản ánh được sự biến thiên giữa người trả lời.

| Unidimensionality | Các item trong thang đo chỉ phản ánh một nhân tố duy nhất | Cần kiểm tra bằng Exploratory Factor Analysis (EFA), không phải chỉ bằng Cronbach's Alpha |
| Cronbach's Alpha | Chỉ số kiểm tra các câu hỏi trong cùng nhóm có ăn khớp với nhau không | Alpha của `Enjoyment` hoặc `Urgency` |
| Corrected Item-Total Correlation | Chỉ số cho biết một câu hỏi có biến thiên cùng chiều với tổng điểm của phần còn lại hay không | Nếu một câu trong `Enjoyment` có chỉ số thấp, câu đó có thể đang lệch với 4 câu còn lại |
| Cronbach's Alpha if Item Deleted | Chỉ số cho biết nếu bỏ một câu hỏi thì alpha của cả thang đo sẽ thay đổi thế nào | Nếu bỏ một câu mà alpha tăng mạnh, câu đó có thể cần xem lại |
| Mã hóa ngược chiều | Câu hỏi có ý nghĩa ngược với các câu còn lại và cần đảo mã trước khi phân tích | Nếu có câu kiểu "Tôi không thấy áp lực khi có flash sale" |

## Sơ Đồ Tóm Tắt Nhanh

![Flow đọc kết quả Cronbach's Alpha](assets/cronbach_alpha_flow_final.svg)

Sơ đồ này dùng để nhìn nhanh thứ tự kiểm tra. Khi diễn giải chi tiết, vẫn nên bám vào phần `Cách đọc kết quả` ở phía trên.


```python
import pyreadstat

# Đọc bộ dữ liệu đã được làm sạch và lưu ở định dạng .sav.
df, meta = pyreadstat.read_sav('data/Survey result_Flash sale.sav')
print(f'Số dòng: {df.shape[0]}, Số cột: {df.shape[1]}')
df.head()
```

```python
import pandas as pd


def prepare_items(items_df):
    return items_df.apply(pd.to_numeric, errors='coerce').dropna()


def cronbach_alpha(items_df):
    k = items_df.shape[1]
    if k < 2:
        raise ValueError("Cronbach's Alpha requires at least two items.")

    item_variances = items_df.var(axis=0, ddof=1).sum()
    total_variance = items_df.sum(axis=1).var(ddof=1)
    if pd.isna(total_variance) or total_variance == 0:
        raise ValueError("Total score variance must be greater than 0 to compute Cronbach's Alpha.")

    return (k / (k - 1)) * (1 - item_variances / total_variance)


def interpret_alpha(alpha):
    if alpha >= 0.90:
        return 'Rất tốt'
    if alpha >= 0.80:
        return 'Tốt'
    if alpha >= 0.70:
        return 'Chấp nhận được'
    if alpha >= 0.60:
        return 'Hơi yếu, cần xem lại'
    return 'Yếu, cần điều chỉnh'


scales = {
    'Eco_Benefit': [f'Eco_Benefit_{i}' for i in range(1, 6)],
    'Enjoyment': [f'Enjoyment_{i}' for i in range(1, 6)],
    'Impulse': [f'Impulse_{i}' for i in range(1, 6)],
    'Qty_Scarcity': [f'Qty_Scarcity_{i}' for i in range(1, 6)],
    'Social_Proof': [f'Social_Proof_{i}' for i in range(1, 6)],
    'Urgency': [f'Urgency_{i}' for i in range(1, 6)],
    'Visual': [f'Visual_{i}' for i in range(1, 6)],
    'Voucher_Scarcity': [f'Voucher_Scarcity_{i}' for i in range(1, 6)],
    'Time_Limit': ['Limit_Time_1', 'Limit_Time_2', 'Time_Limit_3', 'Time_Limit_4', 'Time_Limit_5'],
}

results = []

for scale_name, columns in scales.items():
    scale_df = prepare_items(df[columns])
    alpha = cronbach_alpha(scale_df)
    results.append(
        {
            'Thang đo': scale_name,
            'Số biến': len(columns),
            'Số quan sát hợp lệ': len(scale_df),
            'Cronbach Alpha': round(alpha, 3),
            'Diễn giải': interpret_alpha(alpha),
        }
    )

results_df = pd.DataFrame(results).sort_values('Cronbach Alpha', ascending=False).reset_index(drop=True)
results_df
```
