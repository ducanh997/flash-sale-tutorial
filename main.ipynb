# Tìm Hiểu Về Cronbach's Alpha Từ Dữ Liệu Flash Sale Shopee

---

## 1. Bắt đầu từ một tình huống rất quen với marketer

Giả sử bạn muốn trả lời một câu hỏi rất thực tế:

> **Flash sale có thật sự làm khách hàng thấy gấp, thấy thích thú, hoặc dễ mua ngẫu hứng hơn không?**

Nếu chỉ hỏi một câu, kết quả thu được thường không đủ để phản ánh đầy đủ nội dung cần đo và dễ bị ảnh hưởng bởi cách diễn đạt của riêng câu hỏi đó. Vì vậy, trong nghiên cứu marketing, người ta thường dùng **một nhóm item** để đo cùng một **construct**.

Ví dụ:

- để đo **`Urgency` (cảm giác cấp bách)**, có thể hỏi nhiều câu về việc sợ hết hàng, sợ bỏ lỡ giá tốt, phải quyết định nhanh,
- để đo **`Enjoyment` (cảm xúc tận hưởng)**, có thể hỏi nhiều câu về cảm giác vui vẻ, hạnh phúc, phấn khích,
- để đo **`Impulse Buying`** (trong file dữ liệu được đặt tên ngắn là `Impulse`), có thể hỏi nhiều câu về hành vi mua không có kế hoạch từ trước.

Khi đã có một nhóm item như vậy, ta cần kiểm tra xem chúng có cho thấy **độ nhất quán nội tại đủ tốt trên mẫu hiện tại** để được xem là **bằng chứng reliability sơ bộ** cho construct đó hay không. Đó chính là lúc dùng **Cronbach's Alpha**.

---

## 2. Cronbach's Alpha là gì?

Cronbach's Alpha là một chỉ số giúp trả lời câu hỏi sau:

> **Các item trong cùng một construct có mức độ nhất quán với nhau hay không?**

Hiểu đơn giản:

- **Alpha cao**: các item khá ăn khớp với nhau và cho thấy mức độ nhất quán nội tại tốt trên mẫu hiện tại.
- **Alpha thấp**: các item chưa hợp nhau lắm, có thể có item bị lệch ý, bị mã hóa sai, chịu ảnh hưởng bởi dữ liệu nhiễu, hoặc đang đo sang phần khác.

Bạn có thể hiểu Cronbach's Alpha như một cách kiểm tra **độ nhất quán nội tại** của thang đo. Nói dễ hiểu hơn, nó cho biết các item trong cùng một nhóm có biến thiên đủ đồng thuận để cung cấp **bằng chứng reliability sơ bộ** cho construct đang xét hay không.

Điểm quan trọng cần nhớ:

- Cronbach's Alpha chủ yếu cho biết **mức độ nhất quán giữa các item trong cùng một thang đo**.
- Nó **không đủ để kết luận** thang đo là **hợp lệ**.
- Nó **không tự khẳng định** rằng các item chỉ đo **một khái niệm duy nhất**; điều đó cần được kiểm tra thêm bằng phân tích nhân tố.
- Alpha có thể tăng khi **số lượng item tăng**, nên alpha cao không phải lúc nào cũng đồng nghĩa với chất lượng đo lường tốt hơn.
- Ngay cả khi alpha cao, các item từ nhiều construct khác nhau vẫn có thể có tương quan trong một bối cảnh cụ thể; nếu gộp chúng lại, composite score sẽ trở nên mơ hồ về mặt khái niệm.
- Nếu muốn nhìn kỹ từng item, cần xem thêm các chỉ số như `Corrected Item-Total Correlation` và `Cronbach's Alpha if Item Deleted`.
- Kết quả này luôn phải được hiểu **trên mẫu dữ liệu hiện tại**, không nên xem như một nhãn đúng cho mọi mẫu dữ liệu khác.

Trong quy trình đo lường đầy đủ, Cronbach's Alpha thường được dùng như **bước reliability sơ bộ theo từng construct**. Sau đó, nghiên cứu thường tiếp tục với **Exploratory Factor Analysis (EFA)** để khám phá cấu trúc nhân tố, đánh giá lại reliability trên các factor được giữ lại, và nếu thiết kế nghiên cứu yêu cầu bước xác nhận thì tiếp tục với **Confirmatory Factor Analysis (CFA)**.

## 3. Ví dụ thật từ bộ dữ liệu này

### Ví dụ 1: Thang đo `Urgency`

Trong bộ dữ liệu này, thang đo `Urgency` được đo bằng 5 câu hỏi thật như sau:

- `Urgency_1`: "Tôi cho rằng mình cần phải mua ngay lập tức để sở hữu được sản phẩm này."
- `Urgency_2`: "Tôi cho rằng nếu không mua ngay, sản phẩm sẽ sớm bị bán hết."
- `Urgency_3`: "Tôi cho rằng mình phải quyết định mua nhanh trước khi Flash Sale kết thúc."
- `Urgency_4`: "Tôi cho rằng nếu không mua ngay, tôi có thể bỏ lỡ cơ hội sở hữu sản phẩm với giá ưu đãi."
- `Urgency_5`: "Tôi cho rằng mình sẽ hối tiếc nếu không mua ngay lập tức."

Nếu một người thực sự cảm thấy cấp bách, họ thường sẽ chọn điểm khá cao ở nhiều câu trong nhóm này. Khi đó, các item được xem là khá ăn khớp với nhau.

Trong dữ liệu hiện tại, thang đo `Urgency` có Cronbach's Alpha = `0.876`, tức là **cho thấy bằng chứng reliability sơ bộ tốt trên mẫu hiện tại**.

### Ví dụ 2: Thang đo `Enjoyment`

Thang đo `Enjoyment` trong bộ dữ liệu này gồm 5 câu hỏi:

- `Enjoyment_1`: Flash Sale khiến tôi cảm thấy vui vẻ.
- `Enjoyment_2`: Flash Sale khiến tôi cảm thấy hạnh phúc.
- `Enjoyment_3`: Flash Sale khiến tôi cảm thấy phấn khích.
- `Enjoyment_4`: Flash Sale khiến tôi cảm thấy thích thú.
- `Enjoyment_5`: Flash Sale khiến tôi giảm bớt cảm giác mệt mỏi.

Nếu người trả lời chọn cao ở hầu hết các câu này, điều đó cho thấy nhóm item này có mức độ nhất quán nội tại khá tốt trên mẫu hiện tại. Tuy vậy, item `Enjoyment_5` có sắc thái nội dung hơi khác 4 item còn lại, nên nếu phân tích sâu hơn vẫn nên xem thêm `Corrected Item-Total Correlation`, `Cronbach's Alpha if Item Deleted`, và kết quả EFA.

Trong dữ liệu hiện tại, thang đo `Enjoyment` có Cronbach's Alpha = `0.902`, tức là **cho thấy bằng chứng reliability sơ bộ rất tốt trên mẫu hiện tại**.

### Ví dụ 3: Một trường hợp alpha thấp (ví dụ minh họa)

Ngoài ví dụ từ dữ liệu thật, ta cũng nên xem một ví dụ giả định về trường hợp alpha thấp để dễ phân biệt.

Đây là **ví dụ giả định**, không phải kết quả thật từ bộ dữ liệu Shopee Flash Sale.

Giả sử ta muốn đo `Enjoyment`, nhưng lại gom các câu hỏi sau vào cùng một nhóm:

- "Tôi thấy mua flash sale khá vui."
- "Tôi cảm thấy hào hứng khi tham gia flash sale."
- "Flash sale làm trải nghiệm mua sắm thú vị hơn."
- "Tôi thường đọc đánh giá sản phẩm rất kỹ trước khi mua."
- "Tôi quan tâm đến tốc độ giao hàng hơn là giá khuyến mãi."

Trong nhóm này, 3 câu đầu đang nói về **cảm xúc thích thú**, nhưng 2 câu sau lại nghiêng về **thói quen cân nhắc khi mua hàng**.

Vì các câu hỏi trong nhóm này không đủ nhất quán để hoạt động như một thang đo chung, Cronbach's Alpha có thể xuống thấp, ví dụ chỉ khoảng `0.55`.

Khi gặp trường hợp như vậy, nhà nghiên cứu nên xem lại:

- các câu hỏi có đang đo cùng một construct hay không,
- có item nào bị lệch ý so với phần còn lại hay không,
- có lỗi mã hóa, dữ liệu nhiễu, hoặc đặc điểm mẫu nào đang ảnh hưởng đến kết quả hay không,
- có nên ghi nhận hạn chế hoặc cân nhắc loại bớt item khỏi thang đo hiện tại hay không.

### Ví dụ 4: Câu hỏi đảo chiều nếu không xử lý đúng sẽ làm alpha thấp đi

Đây cũng là **ví dụ giả định**, nhưng rất hay gặp trong nghiên cứu thực tế.

Giả sử bạn có 3 câu cùng đo cảm giác cấp bách:

- "Tôi thấy cần mua ngay khi có flash sale."
- "Tôi sợ hết hàng nếu không mua ngay."
- "Tôi không thấy gấp chút nào khi có flash sale."

Một người thật sự thấy rất cấp bách có thể trả lời như sau:

- câu 1: `5`
- câu 2: `5`
- câu 3: `1`

Về mặt ý nghĩa, câu trả lời này rất hợp lý. Người đó đồng ý mạnh với 2 câu đầu và không đồng ý với câu thứ 3 vì câu thứ 3 đang viết theo hướng ngược lại.

Nhưng nếu giữ nguyên dữ liệu là `5, 5, 1`, phần mềm sẽ dễ hiểu rằng câu thứ 3 đang đi ngược với 2 câu còn lại. Khi đó Cronbach's Alpha có thể bị kéo xuống thấp một cách oan uổng.

Lúc này cần **đảo mã** câu hỏi ngược chiều trước khi tính alpha:

- `1` đổi thành `5`
- `2` đổi thành `4`
- `3` giữ nguyên `3`
- `4` đổi thành `2`
- `5` đổi thành `1`

Sau khi đảo mã, bộ điểm `5, 5, 1` sẽ trở thành `5, 5, 5`. Khi đó phần mềm mới hiểu đúng rằng cả 3 câu đều đang phản ánh cùng một mức độ cấp bách.

Vì vậy, khi thấy alpha thấp, đừng vội kết luận thang đo kém. Trước tiên nên kiểm tra xem có câu hỏi đảo chiều mà chưa được xử lý hay không.

### Ví dụ 5: Nếu mẫu quá nhỏ thì alpha gần như không nói được nhiều

Đây cũng là một tình huống rất quan trọng nhưng dễ bị bỏ qua.

Giả sử bạn có một thang đo gồm 5 câu hỏi, nhưng chỉ có **1 người** trả lời khảo sát.

Người đó trả lời như sau:

- câu 1: `5`
- câu 2: `5`
- câu 3: `4`
- câu 4: `5`
- câu 5: `4`

Nhìn bằng mắt thường, bạn có thể thấy người này trả lời khá nhất quán. Nhưng vấn đề là **Cronbach's Alpha không được tạo ra để đọc từ 1 người trả lời**.

Lý do là alpha cần nhìn sự biến thiên giữa **nhiều người** để xem các câu hỏi có cùng thay đổi với nhau hay không. Nếu chỉ có 1 dòng dữ liệu, gần như không có đủ thông tin để tính phương sai và hiệp phương sai một cách có ý nghĩa.

Nói dễ hiểu hơn:

- `1 người trả lời` thì không thể kết luận thang đo tốt hay không tốt bằng Cronbach's Alpha,
- `rất ít người trả lời` thì dù phần mềm có tính ra một con số, con số đó cũng rất dễ bất ổn,
- mẫu càng đủ lớn và đa dạng, việc diễn giải alpha mới càng đáng tin hơn.

Vì vậy, khi đọc alpha, luôn nên tự hỏi thêm một câu:

> **Con số này được tính trên một mẫu đủ lớn để tin được hay chưa?**

### Ví dụ 6: Alpha cao nhưng không nên gộp khi item đến từ nhiều construct khác nhau

Đây là **ví dụ giả định**, nhưng phản ánh một rủi ro thực tế thường gặp trong nghiên cứu hành vi tiêu dùng.

Giả sử nhà nghiên cứu muốn tóm tắt "tác động tâm lý tổng thể của flash sale" bằng một điểm số chung, và gộp các item sau vào cùng một thang đo:

- "Tôi cảm thấy cần phải mua ngay." (đo `Urgency`)
- "Tôi cảm thấy phấn khích khi thấy flash sale." (đo `Enjoyment`)
- "Tôi thấy mức giảm giá rất hấp dẫn." (đo `Eco_Benefit`)

Trong bối cảnh flash sale, ba item này có thể có tương quan khá tốt với nhau: người cảm thấy gấp thường cũng cảm thấy phấn khích, và cũng thấy giá hấp dẫn. Kết quả là Cronbach's Alpha hoàn toàn có thể đạt `0.78–0.82`.

Tuy nhiên, nếu gộp chúng lại thành một điểm số chung, câu hỏi đặt ra là: **điểm đó đang đo cái gì?**

- Không phải `Urgency` thuần túy.
- Không phải `Enjoyment` thuần túy.
- Không phải `Eco_Benefit` thuần túy.

Thang đo trở nên mơ hồ về mặt khái niệm, dù alpha vẫn cao. Đây chính là ranh giới giữa hai câu hỏi khác nhau:

| Công cụ | Câu hỏi đặt ra |
|---|---|
| **Cronbach's Alpha** | Các item này có tương quan đủ tốt với nhau không? |
| **Exploratory Factor Analysis (EFA)** | Chúng có thực sự phản ánh cùng một nhân tố không? |

Alpha cao là điều kiện **cần**, nhưng chưa **đủ**. Nếu EFA cho thấy các item tải lên nhiều nhân tố khác nhau, thì gộp chúng lại thành một thang đo là không phù hợp về mặt học thuật, dù alpha trông ổn.

Nói ngắn gọn: **alpha cao cho phép tính composite score về mặt số học, nhưng không đảm bảo rằng composite score đó có ý nghĩa về mặt khái niệm**.

## 4. Cách đọc kết quả

Các ngưỡng dưới đây chỉ mang tính hướng dẫn tham khảo; việc diễn giải vẫn cần đặt trong bối cảnh nghiên cứu, nội dung item, đặc điểm mẫu và kết quả phân tích nhân tố.

| Giá trị alpha | Có thể hiểu nhanh | Nên làm gì? |
|---|---|---|
| > 0.95 | Rất cao, cần kiểm tra lại | Có thể có item trùng lặp hoặc construct quá hẹp; chưa nên kết luận chỉ từ alpha |
| 0.90-0.95 | Rất tốt trên mẫu hiện tại | Có thể xem là bằng chứng reliability sơ bộ mạnh, nhưng vẫn cần đọc item và kiểm tra cấu trúc |
| 0.80-0.89 | Tốt trên mẫu hiện tại | Thường là tín hiệu khả quan cho bước reliability sơ bộ; vẫn nên xem thêm item-level diagnostics |
| 0.70-0.79 | Chấp nhận được sơ bộ | Có thể chấp nhận ở nhiều nghiên cứu khám phá, nhưng nên đọc kỹ nội dung item trước khi đi tiếp |
| 0.60-0.69 | Mức cân nhắc | Cần xem lại nội dung câu hỏi, cách mã hóa, dữ liệu và đặc điểm mẫu |
| < 0.60 | Thấp | Chưa nên yên tâm; cần kiểm tra lại thang đo trước |

Nếu muốn kiểm tra kỹ hơn sau khi nhìn alpha, có 5 bước nên làm theo thứ tự:

1. **Đọc lại nội dung item**: xem có item nào bị viết lệch ý hoặc quá giống item khác không.
2. **Kiểm tra mã hóa và câu hỏi đảo chiều**: đảm bảo item ngược chiều đã được recode đúng.
3. **Xem `Corrected Item-Total Correlation`**: biết có item nào không đi cùng phần còn lại của thang đo không.
4. **Xem `Cronbach's Alpha if Item Deleted`**: biết nếu bỏ một item đi thì cả thang đo có tốt hơn rõ rệt không.
5. **Cân nhắc lý thuyết, dữ liệu và đặc điểm mẫu trước khi quyết định loại item**: không nên loại item chỉ vì số liệu nếu item đó là cốt lõi của construct.

Ví dụ:

- Nếu `Urgency = 0.876`, có thể nói thang đo này cho thấy **bằng chứng reliability sơ bộ tốt** trên mẫu hiện tại.
- Nếu `Enjoyment = 0.902`, có thể nói thang đo này cho thấy **bằng chứng reliability sơ bộ rất tốt** trên mẫu hiện tại; nếu muốn kiểm tra kỹ hơn, hãy xem có item nào yếu hơn phần còn lại hay không.
- Nếu một thang đo chỉ đạt `0.58`, chưa nên dùng ngay như một thang đo ổn định. Trước tiên hãy kiểm tra lại nội dung item, cách mã hóa và dữ liệu.

---

## 5. Khi nào nên dùng Cronbach's Alpha?

Cronbach's Alpha phù hợp khi:

- **một construct** được đo bằng **nhiều item Likert**,
- các item được **nhóm theo lý thuyết** thành từng thang đo riêng,
- các item được **dự kiến tổng hợp thành một điểm số chung** cho cùng một construct,
- các item được mã hóa theo **cùng một chiều** (hoặc đã đảo mã câu ngược chiều).

Cronbach's Alpha không phù hợp khi:

- chỉ có một câu hỏi,
- đó là thông tin mô tả mẫu như giới tính, tuổi, nghề nghiệp,
- các item trong cùng nhóm thực ra đang đo nhiều construct khác nhau,
- số quan sát quá ít để phản ánh được sự biến thiên giữa người trả lời.

## 6. Bảng giải thích thuật ngữ

| Thuật ngữ | Nghĩa đơn giản | Ví dụ trong bộ dữ liệu này |
|---|---|---|
| Khái niệm (`construct`) | Nội dung mà nghiên cứu muốn đo | `Urgency`, `Enjoyment`, `Impulse Buying` |
| Biến quan sát (`item`) | Một câu hỏi cụ thể trong bảng khảo sát | `Urgency_1`, `Urgency_2` |
| Thang đo (`scale`) | Một nhóm item cùng đo một construct | 5 item của `Urgency` |
| Thang Likert | Câu hỏi có trả lời theo mức độ đồng ý | `1 = Rất không đồng ý` đến `5 = Rất đồng ý` |
| Độ nhất quán nội tại (`internal consistency`) | Mức độ các item trong cùng nhóm có biến thiên đồng thuận với nhau hay không | Các item của `Enjoyment` có xu hướng cùng tăng hoặc cùng giảm khi người trả lời thay đổi mức độ cảm xúc |
| Tính hợp lệ (`validity`) | Thang đo có thật sự đo đúng điều cần đo không | `Urgency` có thật sự đo cảm giác cấp bách không |
| Tính đơn hướng (`unidimensionality`) | Các item trong thang đo chỉ phản ánh một nhân tố duy nhất | Cần kiểm tra bằng phân tích nhân tố, không phải chỉ bằng Cronbach's Alpha |
| Cronbach's Alpha | Chỉ số kiểm tra các item trong cùng nhóm có ăn khớp với nhau không | Alpha của `Enjoyment` hoặc `Urgency` |
| Corrected Item-Total Correlation | Chỉ số cho biết một item có biến thiên cùng chiều với tổng điểm của phần còn lại hay không | Nếu một item trong `Enjoyment` có chỉ số thấp, item đó có thể đang lệch với 4 item còn lại |
| Cronbach's Alpha if Item Deleted | Chỉ số cho biết nếu bỏ một item thì alpha của cả thang đo sẽ thay đổi thế nào | Nếu bỏ một item mà alpha tăng mạnh, item đó có thể cần xem lại |
| Mã hóa ngược chiều (`reverse coding`) | Item có ý nghĩa ngược với các item còn lại và cần đảo mã trước khi phân tích | Nếu có item kiểu "Tôi không thấy áp lực khi có flash sale" |

## Sơ Đồ Tóm Tắt Nhanh

![Flow đọc kết quả Cronbach's Alpha](assets/cronbach_alpha_flow_final.svg)

Sơ đồ này tóm tắt **workflow measurement** rộng hơn. Notebook hiện chỉ minh họa bước **Cronbach's Alpha** và cách diễn giải reliability sơ bộ; các bước `EFA`/`CFA` được nhắc đến để đặt Cronbach's Alpha vào đúng bối cảnh phương pháp.


```python
import pyreadstat

# Đọc bộ dữ liệu đã được làm sạch và lưu ở định dạng .sav.
df, meta = pyreadstat.read_sav('data/Survey result_Flash sale.sav')
print(f'Số dòng: {df.shape[0]}, Số cột: {df.shape[1]}')
df.head()
```

```python
import pandas as pd


def prepare_items(items_df):
    return items_df.apply(pd.to_numeric, errors='coerce').dropna()


def cronbach_alpha(items_df):
    k = items_df.shape[1]
    if k < 2:
        raise ValueError("Cronbach's Alpha requires at least two items.")

    item_variances = items_df.var(axis=0, ddof=1).sum()
    total_variance = items_df.sum(axis=1).var(ddof=1)
    if pd.isna(total_variance) or total_variance == 0:
        raise ValueError("Total score variance must be greater than 0 to compute Cronbach's Alpha.")

    return (k / (k - 1)) * (1 - item_variances / total_variance)


def interpret_alpha(alpha):
    if alpha > 0.95:
        return 'Rất cao, cần kiểm tra lại'
    if alpha >= 0.90:
        return 'Rất tốt trên mẫu hiện tại'
    if alpha >= 0.80:
        return 'Tốt trên mẫu hiện tại'
    if alpha >= 0.70:
        return 'Chấp nhận được sơ bộ'
    if alpha >= 0.60:
        return 'Mức cân nhắc'
    return 'Thấp, cần xem lại'


construct_labels = {
    'Eco_Benefit': 'Nhận thức về lợi ích kinh tế',
    'Enjoyment': 'Cảm xúc tận hưởng',
    'Impulse': 'Hành vi mua sắm ngẫu hứng (Impulse Buying)',
    'Qty_Scarcity': 'Nhận thức về giới hạn số lượng sản phẩm',
    'Social_Proof': 'Nhận thức về bằng chứng xã hội',
    'Urgency': 'Cảm giác cấp bách',
    'Visual': 'Nhận thức về khả năng hiển thị',
    'Voucher_Scarcity': 'Nhận thức về giới hạn số lượng voucher',
    'Time_Limit': 'Nhận thức về giới hạn thời gian',
}

scales = {
    'Eco_Benefit': [f'Eco_Benefit_{i}' for i in range(1, 6)],
    'Enjoyment': [f'Enjoyment_{i}' for i in range(1, 6)],
    'Impulse': [f'Impulse_{i}' for i in range(1, 6)],
    'Qty_Scarcity': [f'Qty_Scarcity_{i}' for i in range(1, 6)],
    'Social_Proof': [f'Social_Proof_{i}' for i in range(1, 6)],
    'Urgency': [f'Urgency_{i}' for i in range(1, 6)],
    'Visual': [f'Visual_{i}' for i in range(1, 6)],
    'Voucher_Scarcity': [f'Voucher_Scarcity_{i}' for i in range(1, 6)],
    'Time_Limit': ['Limit_Time_1', 'Limit_Time_2', 'Time_Limit_3', 'Time_Limit_4', 'Time_Limit_5'],
}

results = []

for scale_name, columns in scales.items():
    scale_df = prepare_items(df[columns])
    alpha = cronbach_alpha(scale_df)
    results.append(
        {
            'Mã thang đo': scale_name,
            'Khái niệm / construct': construct_labels[scale_name],
            'Số biến': len(columns),
            'Số quan sát hợp lệ': len(scale_df),
            'Cronbach Alpha': round(alpha, 3),
            'Diễn giải sơ bộ': interpret_alpha(alpha),
        }
    )

results_df = pd.DataFrame(results).sort_values('Cronbach Alpha', ascending=False).reset_index(drop=True)
results_df
```
